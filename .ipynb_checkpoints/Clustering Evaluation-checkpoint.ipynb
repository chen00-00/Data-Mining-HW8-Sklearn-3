```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering

df = pd.read_csv("data.csv")
data = df.drop(['filename', 'label'],axis=1)
data
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score

k_means = KMeans(n_clusters=2, max_iter=50, init='k-means++', n_init='auto', random_state=None)
k_means.fit(data)
label1 = k_means.labels_
ac = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='complete')
ac.fit(data)
label2 = ac.labels_
normalized_mutual_info_score(label1, label2)
```

```python
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

X, y = make_blobs(random_state=42)
silhouette_score(X, k_means.fit_predict(X))
```
